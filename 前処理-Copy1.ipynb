```python
'''
# 4分の１サイズの画像生成
files = glob.glob('17ダイコン写真/*.jpg')

for f in files:  
    fname = os.path.basename(f)
    img = Image.open(f)
    img_resize = img.resize((int(img.width/4), int(img.height/4)))
    img_resize.save('qpics/' + fname)
    '''
# データ管理用データフレームを生成する
import os
import glob
import pandas as pd
import re # 正規表現のライブラリ

df = pd.DataFrame(columns =['filename', 'year','spacies', 'block', 'shape','sn'])

files = glob.glob('qpics/*.jpg')

# 表データの生成
for f in files:    
    fname = os.path.basename(f)
    try:  #  ファイル名は、年、株名、区画、シリアル番号から成っているという前提で、これを分離する
        (year,spacies, block, shape,sn,_) = re.findall(r'(\d+|\D+)',fname)  
        sn = ('00'+sn)[-3:] # シリアル番号を 　００１、００２　... とするための処理
    except: # 上記の前提に合わないファイル名の画像が含まれていたら例外
        continue 
    # 分解したデータを行データとしてデータフレームに追加する。
    sd = pd.Series({'filename':fname, 'year':year, 'spacies':spacies, 'block':block, 'shape':shape,'sn':sn })
    df = df.append(sd, ignore_index=True)
    # 並べ替え
    df = df.sort_values(by=['spacies','block','shape','sn'])   
```

### re.findall についてのメモ

re は正規表現を扱うためのライブラリ
findall(PAT, STR)　は 文字列STR 中で、正規表現PAT にマッチする部分文字列のリストを返す。

r'   ' は文字列中のエスケープ文字の効果を無視することを意味する。例えば、 '\n' は通常は改行文字を表すが、 　r'\n'　と書くと
見た通りの、バックスラッシュ文字と　n　からなる２文字の文字列という意味になる。

正規表現　 \d  は数字文字一つ、　\D は非数字文字一つ
＋ はそれが１つ以上連続することを意味し、｜はいわゆる論理和を表す

以上より、r'(\d+|\D+)' は、”数値であるか数字を含まない字文列”　を表すことになる。

findall の返り値はたとえ数値であっても文字列であることに注意が必要。

# 画像からダイコンの根部のみを抽出する

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from PIL import Image

# 画像例
filename = 'qpics/17kazafu2a10.jpg'
pilimg = Image.open(filename)
pilimg.resize((int(pilimg.width/2),int(pilimg.height/2)))
```

```python
img = cv2.imread(filename)     # 画像の読み込み（カラー）
gry = cv2.imread(filename,0)   # 画像の読み込み（グレイ）
height,width = gry.shape    # 入力画像の縦横サイズ
height,width
```

# 根部の抽出

根部とそれ以外を分離したい。方針としては、適当な閾値で２階調化した画像の白連結成分をラベル付けし、面積の大きな領域を抜き出すことで根部を抜き出すという方向で考える。

## 基礎実験
対象がダイコンで、先の画像サンプルのように黒幕を背景として撮影されているならば、単純な２階調化でも良好な結果がえられる可能性が高い。
 
###  中間調 127 で２階調化 

```python
ret, bw = cv2.threshold(gry,127,255,cv2.THRESH_BINARY)
Image.fromarray(bw).resize((int(pilimg.width/2),int(pilimg.height/2)))
```

大津の方法（判別分析法）で２階調化

```python
ret, bw = cv2.threshold(gry,127,255,cv2.THRESH_OTSU)
Image.fromarray(bw).resize((int(pilimg.width/2),int(pilimg.height/2)))
```

```python
#　入力画像のカラー分布を　混合数5の混合ガウス分布で近似する


from sklearn import cluster, preprocessing, mixture #機械学習用のライブラリを利用
gmm=mixture.GaussianMixture(n_components=5,covariance_type='full')    
fimg = img.reshape((height*width,3))
z_gmm=gmm.fit(fimg)
```

```python
z_gmm.means_
```

```python
levels = []
maxlevel = np.linalg.norm([255.0,255.0,255.0])
print()
for m in z_gmm.means_:
    levels.append(np.linalg.norm(m))

levels = np.array(np.sort(levels))
blacklevel = levels[0]   # 黒領域の色ノルムの平均値
whitelevel = levels[-1] # 白領域の色ノルムの平均値

mask0 = np.zeros((height,width),dtype=np.uint8)
for i in range(height):
    for j in range(width):
        if np.linalg.norm(img[i][j]) < blacklevel:
            mask0[i][j] = 255

mask1 = np.zeros((height,width),dtype=np.uint8)
for i in range(height):
    for j in range(width):
        if np.linalg.norm(img[i][j]) > whitelevel:
            mask1[i][j] = 255
            
plt.imshow(mask0),plt.show()
plt.imshow(mask1),plt.show()
```
